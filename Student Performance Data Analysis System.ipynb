```python
# Step 1 : Import Required Libraries
```

```python
# Import required libraries #

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("Libraries imported successfully")
```

```python
# Step 2 : Load Dataset
```

```python
# i) Load CSV dataset #
df = pd.read_csv("E:/Arbor_Acedemy/Pandas/student_data.csv")
print(df)
```

```python
# ii) Display first few rows #
print(df.head())
```

```python
# iii) Display dataset information
print(df.info())
```

```python
# iv) Check number of rows and columns
print(df.shape)
```

```python
# Step 3 : Data Cleaning
```

```python
# i) Check null values
print(df.isnull().sum())
```

```python
# ii) Remove duplicate records
print(df.duplicated().sum())
```

```python
# iii) Verify data types
df = df.drop_duplicates()
```

```python
# iv) Handle missing values if present
print(df.dtypes)
```

```python
# Step 4: Data Analysis
```

```python
# Calculate total marks

df["Total_Marks"] = (df["Maths"] + df["Science"] + df["English"])
df.head()
```

```python
# Calculate percentage
df["Percentage"] = (df["Total_Marks"]/300)*100
(df.head())
```

```python
# Assign grades based on percentage

def assign_grade(p):
    if p >= 90:
        return "A+"
    elif p >= 80:
        return "A"
    elif p >= 70:
        return "B"
    elif p >= 60:
        return "C"
    else:
        return "D"
df["Grade"] = df["Percentage"].apply(assign_grade)
(df)
```

```python
# Step 5: Generate Insights
```

```python
# i) Top performing student
top_student = df.loc[df["Percentage"].idxmax()]

print("\nTop Performing Student\n")

print(top_student)
```

```python
# ii) Lowest performing student
low_student = df.loc[df["Percentage"].idxmin()]

print("\nLowest Performing Student\n")

print(low_student)
```

```python
# iii) Average marks in each subject

print("\nAverage Marks in Each Subject\n")

print(df[["Maths","Science","English"]].mean())
```

```python
# iv) Highest attendance
highest_attendance = df.loc[df["Attendance"].idxmax()]
print("\nHighest Attendance\n")
print(highest_attendance)
```

```python
# v) Students scoring below average

average = df["Percentage"].mean()
below_average = df[df["Percentage"] < average]
print("\nStudents Below Average\n")
print(below_average[["Name","Percentage"]])
```

```python
# vi) Gender-wise average performance
gender_average = (df.groupby("Gender")["Percentage"].mean())
print(gender_average)
```

```python
# Step 6: Visualization (Optional Bonus)
```

```python
# i) Bar chart
plt.figure(figsize=(8,5))
plt.bar(df["Name"],df["Percentage"])
plt.title("Student Percentage")
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.show()
```

```python
# ii) Pie Chart

gender_count = df["Gender"].value_counts()
plt.figure(figsize=(7,7))
plt.pie(gender_count,labels=gender_count.index,autopct="%1.1f%%",startangle=90)
plt.title("Gender Distribution")
plt.show()
```

```python
# iii) Line chart
plt.figure(figsize=(10,5))
plt.plot(df["Name"],df["Attendance"],marker='o',linewidth=2)
plt.title("Student Attendance Comparison")
plt.xlabel("Students")
plt.ylabel("Attendance (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# iv) Attendance comparison chart
plt.figure(figsize=(8,5))
plt.plot(df["Name"],df["Attendance"])
plt.title("Attendance Comparison")
plt.xlabel("Students")
plt.ylabel("Attendance")
plt.show()
```

```python
# Step 7: Export Final Report
# 1) Save processed dataset into new CSV file

df.to_csv("final_report.csv",index=True)
print("\nFinal Report Saved")
print("\nProject Completed Successfully")
```
